```python
import pandas as pd
from collections import defaultdict
results = defaultdict(list)
num_experiments= range(5)

for i in num_experiments:
    with open('../results/mm_results-%d.log' % i,'r') as f:
        for l in f:
            if l.startswith('m='):
    #             print l.strip()
                args = map(lambda a: a.strip().split('=')[1],l.strip().split(' '))
                m = int(args[0])
                k = int(args[1])
                n = int(args[2])
#                 mkn = ','.join(args[:3])
                sparsity = float(args[3])
                iterations = int(args[4])
            elif l.startswith('Time for'):
                args = l.strip().split(' ')[2].strip()
                alg = args.split(':')[0].split('_')[-1]
                part = args.split(':')[1].split('=')[0]
                t = float(args.split(':')[1].split('=')[1])
    #             print l.strip()
                results[(m,k,n,sparsity,iterations,alg,part)].append(t/iterations)

results_avgs = [list(key)+[sum(vals)/len(vals)] for key,vals in sorted(results.items(),key=lambda item:item[0][:3])]
df = pd.DataFrame(results_avgs,columns=['m','k','n','sparsity','iterations','alg','part','t'])
print df
```
